```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('data/VN_House_price_Clean2.csv')
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    x="Floors",
    y="Price",
    data=df
)
plt.title("Price by Number of Floors")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    x="Bedrooms",
    y="Price",
    data=df
)
plt.title("Price by Bedrooms")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(
    x="Bathrooms",
    y="Price",
    data=df
)
plt.title("Price by Bathrooms")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ch ly cc ct s
num_cols = [
    "Area",
    "Frontage",
    "Access Road",
    "Floors",
    "Bedrooms",
    "Bathrooms",
    "Price"
]

# Tnh ma trn tng quan
corr = df[num_cols].corr()

# V heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    square=True
)

plt.title("Correlation Heatmap of Physical Features and Price")
plt.tight_layout()
plt.show()
```
