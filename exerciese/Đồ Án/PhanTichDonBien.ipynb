```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('data/VN_House_price_Clean2.csv')
```

```python
numeric_cols = [
    "Area", "Frontage", "Access Road",
    "Floors", "Bedrooms", "Bathrooms", "Price"
]
def plot_hist_box(df, col, bins=30):
    plt.figure(figsize=(10, 4))

    # Histogram
    plt.subplot(1, 2, 1)
    plt.hist(df[col], bins=bins)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title(f"Histogram of {col}")

    # Boxplot
    plt.subplot(1, 2, 2)
    plt.boxplot(df[col], vert=False)
    plt.xlabel(col)
    plt.title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()


for col in numeric_cols:
    plot_hist_box(df, col)
```
