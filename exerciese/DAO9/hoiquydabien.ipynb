```python
import numpy as np
import pandas as pd
```

```python
# Bảng 2-2
data = np.array([
    [91,    998, 4,  89.8],
    [87,   1256, 7,  90.2],
    [86,    791, 6,  81.1],
    [88,    705, 5,  83.0],
    [92.8, 1089, 3,  90.9],
    [95.2, 1135, 6, 119.0],
    [93.3, 1076, 4,  94.9],
    [97.7, 1198, 7, 132.4]
], dtype=float)
```

```python
X = data[:, :3]      # Temperature, Tourists, SunnyDays
y = data[:, 3]       # Sales
```

```python
# Thêm cột hệ số tự do
X_design = np.hstack([X, np.ones((len(X), 1))])
```

```python
# Tính hệ số
coeffs, _, _, _ = np.linalg.lstsq(X_design, y, rcond=None)
b1, b2, b3, b0 = coeffs

print("He so hoi quy:")
print(f"b1 (Temperature): {b1}")
print(f"b2 (Tourists): {b2}")
print(f"b3 (SunnyDays): {b3}")
print(f"Intercept b0: {b0}")
```

```python
# Predicted values
y_pred_multi = X_design @ coeffs

# Error^2
errors_multi = (y - y_pred_multi)**2

# SSE
sse_multi = errors_multi.sum()
```

```python
df_multi = pd.DataFrame({
    "Temp": X[:, 0],
    "Tourists": X[:, 1],
    "SunnyDays": X[:, 2],
    "Sales": y,
    "Predicted": y_pred_multi,
    "Error^2": errors_multi
})

print("\n-- Bang du doan tung dong --")
print(df_multi)

print(f"\nSSE (Hồi quy đa biến) = {sse_multi}")
```
