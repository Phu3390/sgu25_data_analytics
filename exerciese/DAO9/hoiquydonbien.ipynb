```python
import numpy as np
import pandas as pd

temperature = np.array([91, 87, 86, 88, 92.8, 95.2, 93.3, 97.7], dtype=float)
sales = np.array([89.8, 90.2, 81.1, 83.0, 90.9, 119.0, 94.9, 132.4], dtype=float)
```

```python
# Tính m và b
A = np.vstack([temperature, np.ones(len(temperature))]).T
m, b = np.linalg.lstsq(A, sales, rcond=None)[0]

print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
```

```python
# Dự đoán khi Temperature = 88.8
x_new = 88.8
y_pred = m * x_new + b
print(f"\nGia tri du doan tai X = 88.8: {y_pred}")
```

```python
# Dự đoán tất cả
predictions = m * temperature + b

# Lỗi từng điểm
errors = (sales - predictions)**2

# Tổng SSE
sse = errors.sum()
```

```python
df_single = pd.DataFrame({
    "Temperature": temperature,
    "Sales": sales,
    "Predicted": predictions,
    "Error^2": errors
})

print("\n-- Bang du doan tung dong --")
print(df_single)

print(f"\nSSE (Hồi quy đơn) = {sse}")
```
